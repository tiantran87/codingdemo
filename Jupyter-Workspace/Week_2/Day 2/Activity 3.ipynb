```python
# @TODO: Initialize a dictionary of banks and market caps (in billions)
market_cap={
    "Toronto-Dominion Bank": 302,
    "Royal Bank of Canada": 173,
    "Wells Fargo": 273,
    "Goldman Sachs": 87,
    "Morgan Stanley": 72,
    "Canadian Imperial Bank of Commerce": 83,
    "TD Bank": 108,
    "Bank of Montreal": 67,
    "Capital One": 47,
    "National Bank of Canada": 327,
    "FNB Corporation": 4,
    "Laurentian Bank of Canada": 3,
    "Ally Financial": 12,
    "Montreal Trust Company": 145,
    "Canadian Western Bank": 0.97,
}

total=0

# @TODO: Change the market cap for 'Royal Bank of Canada'
#Change the market cap for `Royal Bank of Canada` to `170`.
market_cap["Royal Bank of Canada"]=170
print("New market cap for Royal Bank of Canada:")
print(market_cap["Royal Bank of Canada"])

# @TODO: Add a new bank and market cap pair
#Add a new bank `Scotiabank` to the dictionary and set the market cap to `33`

market_cap["Scotiabank"]=33
print("New market cap for Scotiabank:")
print(market_cap["Scotiabank"])

# @TODO: Remove a bank from the dictionary
#Delete `Montreal Trust Company` from the dictionary, as it is a depreciated bank acquired by Scotiabank in 1994.

market_cap.pop("Montreal Trust Company")
print(market_cap)

# Total market capitalization

for cap in market_cap:
    total = total + market_cap[cap]

#Total market capitalization
print(f"Total market cap: {total}")

#Total number of banks
print(f"Total number of banks: {len(market_cap)}")

#Average market capitalization
avg = (total/len(market_cap))
print(f"Average market capitalization: {round(avg, 2)}")

#Largest bank

largest_cap=0
largest_bank="Z"

for bank, cap in market_cap.items():
    if cap > largest_cap:
        largest_cap = cap
        largest_bank = bank
        
            
print(f"Largest bank is: {largest_bank}  with capital (in billion) {largest_cap}")

#Smallest bank

smallest_cap=0
smallest_bank="Z"

for bank, cap in market_cap.items():
    if cap < smallest_cap or smallest_cap == 0:
        smallest_cap = cap
        smallest_bank = bank
    
print(f"Smallest bank is: {smallest_bank}  with capital (in billion) {smallest_cap}")

# mega-caps`: Market capitalization greater than or equal to `$300 billion`.
mega_cap={}
large_cap={}
mid_cap={}
small_cap={}
for bank, cap in market_cap.items():
    if cap >= 300:
        mega_cap[bank]=cap
    elif cap >= 10 and cap < 300:
        large_cap[bank]=cap
    elif cap >= 2 and cap < 10:
        mid_cap[bank]=cap
    else:
        small_cap[bank]=cap
        
print(f"Mega capital banks: {mega_cap}")
print()
print(f"Large capital banks: {large_cap}")
print()
print(f"Mid capital banks: {mid_cap}")
print()
print(f"Small capital banks: {small_cap}")
print()
```
